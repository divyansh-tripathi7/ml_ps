Ad Sale prediction using logistic regression 


```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('custData.csv')
```

```python
print(dataset.shape)
print(dataset.head(11))
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X,Y,test_size= 0.25, random_state = 0)
```

```python
## feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state= 0)
model.fit(X_train,y_train)
```

```python
##prediction 
```

```python
age = int(input("Enter new customer Age: "))
sal = int(input("enter new customer's salary : "))
newCust = [[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result == 1:
  print(" customer will buy the  product")
if result == 0:
  print(" customer will not buy the  product")
  
```

```python
#confusion matrix
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# evaluating model
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)

print("accuracy of model : {0}%".format(accuracy_score(y_test,y_pred)*100))
```
